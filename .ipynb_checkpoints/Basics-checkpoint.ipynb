# Visualizing Earnings Based on College Majors

```python
%matplotlib inline 
#to display plots inline
```

## Exploring and Cleaning Dataset

```python
import pandas
recent_grads = pandas.read_csv("recent-grads.csv")
print(recent_grads.iloc[0], '/n')
print(recent_grads.head(),'/n')
print(recent_grads.tail())
```

```python
recent_grads.describe()
```

```python
#dropping missing values
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print('How many rows were dropped due to missing values?')
print(raw_data_count - cleaned_data_count)
```

## Visualizations

```python
recent_grads.plot('Sample_size', 'Median', 'scatter', title = 'Sample Size vs. Median')
```

```python
recent_grads.plot('Sample_size', 'Unemployment_rate', 'scatter', title = 'Sample Size vs. Unemployment Rate')
```

```python
recent_grads.plot('Full_time', 'Median', 'scatter', title = 'Full Time Employees vs. Median')
```

```python
recent_grads.plot('ShareWomen', 'Unemployment_rate', 'scatter', title = 'Women Percent Share vs. Unemployment Rate')
```

```python
recent_grads.plot('Men', 'Median', 'scatter', title = 'Amount of Men vs. Median')
```

```python
recent_grads.plot('Women', 'Median', 'scatter', title = 'Amount of Women vs. Median')
```

It seems like the majors that make up the extremeties of median salary as well as umemployment rate have low sample sizes.  There seems to be no correlation between share of women and median salary.

### Histograms

```python
#import matplotlib.pyplot as plt #doesn't seem necessary to import this module
# to be able to use the methods of matplotlib.pyplot's axes objects
fig1, ax1 = plt.subplots()
recent_grads['Sample_size'].hist(bins = 20, range =(0,500), ax = ax1)
ax1.set_xlabel('Sample Size of Major')
```

```python
fig2, ax2 = plt.subplots()
recent_grads['Median'].hist(bins = 20, ax = ax2)
ax2.set_xlabel('Median Salaries')
```

```python
fig3, ax3 = plt.subplots()
print("Number Employed")
recent_grads['Employed'].hist(bins = 20, ax = ax3)
ax3.set_xlabel('Number Employed')
```

```python
fig4, ax4 = plt.subplots()
recent_grads['Full_time'].hist(bins = 20, ax = ax4)
ax4.set_xlabel('Number Working Full Time')
```

```python
fig5, ax5 = plt.subplots()
recent_grads['ShareWomen'].hist(bins = 20, ax = ax5)
ax5.set_xlabel('Women as Share of Total')
```

```python
fig6, ax6 = plt.subplots()
recent_grads['Unemployment_rate'].hist(bins = 18, ax = ax6)
ax6.set_xlabel('Unemployment Rate')
```

```python
fig7, ax7 = plt.subplots()
recent_grads['Men'].hist(bins = 20, range = (0,320000), ax = ax7)
ax7.set_xlabel('Male Graduates')
```

```python
fig8, ax8 = plt.subplots()
recent_grads['Women'].hist(bins = 20, range= (0, 320000), ax = ax8)
ax8.set_xlabel('Female Graduates')
```

Judging from the above histograms, the most common median salary range falls between \$30,000 and \$40,000. There appears to only be five majors that have unemployment rates of about 12 percent or higher.  Most majors appear to have a higher share of women then men.  

### Scatter Matrix Plots

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize = (10, 10))
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize = (15, 15))
```

It can again be seen in the scatter matrix plots above that majors with low sample sizes seem to make up the extremeties of median salary and unemployment rate. 

### Comparing Share of Women vs. Pay

```python
recent_grads.sort_values(by = 'Median')[:10].plot.bar('Median', 'ShareWomen')
recent_grads.sort_values(by = 'Median', ascending = False)[:10].plot.bar('Median', 'ShareWomen')
```

Only one of the top 10 paying majors is predomenantly made up of women, whereas all of the bottom 10 paying majors are predomenantly made up of women.

### Comparing Unemployment Rate Vs. Pay

```python
recent_grads.sort_values(by = 'Median')[:10].plot.bar('Median', 'Unemployment_rate')
recent_grads.sort_values(by = 'Median', ascending = False)[:10].plot.bar('Median', 'Unemployment_rate')
```

In general, it appears that the top 10 paying majors had lower unemployment rates than the lowest paying majors
