# Visualizing Earnings Based on College Majors

```python
%matplotlib inline # to display plots inline
```

## Exploring and Cleaning Dataset

```python
import pandas
recent_grads = pandas.read_csv("recent-grads.csv")
print(recent_grads.iloc[0], '/n')
print(recent_grads.head(),'/n')
print(recent_grads.tail())
```

```python
recent_grads.describe()
```

```python
#dropping missing values
raw_data_count = recent_grads.shape[0]
recent_grads = recent_grads.dropna()
cleaned_data_count = recent_grads.shape[0]
print('How many rows were dropped due to missing values?')
print(raw_data_count - cleaned_data_count)
```

## Visualizations

```python
recent_grads.plot('Sample_size', 'Median', 'scatter', title = 'Sample Size vs. Median')
```

```python
recent_grads.plot('Sample_size', 'Unemployment_rate', 'scatter', title = 'Sample Size vs. Unemployment Rate')
```

```python
recent_grads.plot('Full_time', 'Median', 'scatter', title = 'Full Time Employees vs. Median')
```

```python
recent_grads.plot('ShareWomen', 'Unemployment_rate', 'scatter', title = 'Women Percent Share vs. Unemployment Rate')
```

```python
recent_grads.plot('Men', 'Median', 'scatter', title = 'Amount of Men vs. Median')
```

```python
recent_grads.plot('Women', 'Median', 'scatter', title = 'Amount of Women vs. Median')
```

There does not seem to be a correlation between sample size of people who majored in a particular major and median salary or unemployment rate.  There also seems to be no correlation between share of women or number of full time employees and median salary.
